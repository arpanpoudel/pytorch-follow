```python
def longestSubarray(arr):
    # Write your code here
    max_length=0
    for i in arr:
        diff_arr=[x-i for x in arr]
        if diff_arr.count(1)> diff_arr.count(-1):
            count=diff_arr.count(1)+diff_arr.count(0)
        elif diff_arr.count(1)< diff_arr.count(-1):
            count=diff_arr.count(-1)+diff_arr.count(0)
        else:
            if diff_arr.count(1)>= diff_arr.count(-1):
                count=diff_arr.count(1)
            else:
                count=diff_arr.count(-1)
        print(count)
        if count>max_length:
            max_length=count
    return max_length
```

```python
def rotLeft(a, d):
    for i in range(d):
        for k in range(len(a)):
            if k==0:
                temp=a[0]
                a[0]=a[1]
            elif k!= len(a)-1:
                a[k]=a[k+1]
            else:
                a[k]=temp
    return a
```

```python
print(rotLeft([1,2,3],5))
```
